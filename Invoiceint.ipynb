```python
import easyocr
```

```python
reader = easyocr.Reader(['en'], gpu = True)
```

```python
import PIL
from PIL import ImageDraw
imaginv = PIL.Image.open("/content/drive/MyDrive/invoice/000038.jpg")
imaginv
```

```python
bounds = reader.readtext('/content/drive/MyDrive/invoice/000038.jpg')
bounds
```

```python
def draw_boxes(image, bounds, color='red', width=2):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return image

draw_boxes(imaginv,bounds)
```
